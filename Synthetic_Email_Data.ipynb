# 1. Creating the Synthetic Data

```python
import pandas as pd
import numpy as np
import os

# Set random seed for reproducibility
np.random.seed(42)

# Number of synthetic subscribers
n_subscribers = 10000

# Generate synthetic data
data = {
    'subscriber_id': np.arange(1, n_subscribers + 1),
    'email_frequency_per_week': np.random.randint(1, 10, n_subscribers),
    'content_relevance_score': np.random.uniform(1, 5, n_subscribers),
    'email_send_hour': np.random.randint(0, 24, n_subscribers),
    'personalized': np.random.choice([0, 1], n_subscribers, p=[0.4, 0.6]),
    #'spam_complaints': np.random.choice([0, 1], n_subscribers, p=[0.9, 0.1]),
    'time_in_business': np.random.randint(1, 60, n_subscribers),
    #'competing_priorities_score': np.random.uniform(1, 5, n_subscribers),
    'engagement_score': np.random.uniform(0, 1, n_subscribers),
    'unsubscribe': np.random.choice([0, 1], n_subscribers, p=[0.8, 0.2])
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Adjust engagement score based on factors
df['engagement_score'] = (
    df['engagement_score'] +
    0.1 * (df['content_relevance_score'] - 3) -  # Higher relevance increases engagement
    0.05 * df['email_frequency_per_week'] +     # Higher frequency slightly decreases engagement
    0.2 * df['personalized'] -                  # Personalization boosts engagement
    #0.3 * df['spam_complaints'] -               # Spam complaints decrease engagement
    0.01 * (df['time_in_business'] - 30).abs() / 30   # Length of subscription adjustment
   # 0.1 * (3 - df['competing_priorities_score'])  # Lower competing priorities increase engagement
)

# Normalize engagement score to [0, 1]
df['engagement_score'] = (df['engagement_score'] - df['engagement_score'].min()) / (
    df['engagement_score'].max() - df['engagement_score'].min()
)

# Adjust unsubscribe based on engagement score
df['unsubscribe'] = np.where(df['engagement_score'] < 0.3, 1, df['unsubscribe'])

# Define the folder path
folder_path = r'C:\Users\james\Documents\Capstone\Final'

# Save to a CSV file (optional)
df.to_csv(os.path.join(folder_path, 'synthetic_email_data.csv'), index=False)
```
